```python
import numpy as np
import pandas as pd
import datetime
```

```python
R = np.matrix(np.zeros([2, 2]))
Q = np.matrix(np.zeros([2, 2]))
```

```python
def RMatrix(m, n):
  index=0;
  if(m>1):
   R[index, 1] = (m)*100
  if(m==1):
   R[index, 0] = -2
   index=1
  if(n==1):
    R[index, 1] = +100
  if(n==0):
    R[index, 1] = -1
  
  return R

```

```python
action = [0, 1]
gamma = 1
initial_state=0;
```

```python
df = pd.read_csv('/content/selfish-miner.csv')
```

```python
prev=df.loc[0,'Hash']
date_time_str=df.loc[0,'Timestamp']
date_time_obj =pd.to_datetime(date_time_str)
prevt=datetime.datetime.timestamp(date_time_obj)
start=1;
t=0
while(start<82):
  t=0
  count=1;
  if(df.loc[start,'Parent Hash']!=prev):
    count=1;
    start+1;
    if(prevt>= datetime.datetime.timestamp(pd.to_datetime(df.loc[start,'Timestamp']))):
      t=1;
    while((df.loc[start,'Hash']==prev) and (start<82)):
      count+1
      prev=df.loc[start,'Hash']
      start = start+1
    R1 = RMatrix(count, t)
    
    i=0;
    
    Q[i, action[0]] +=  R1[i, action[0]] + gamma * max(Q[i+1, action[0]], Q[i+1, action[1]])
    Q[i, action[1]] +=  R1[i, action[1]] + gamma * max(Q[i+1, action[0]], Q[i+1, action[1]])

    i = i + 1
    Q[i, action[0]] +=  R1[i, action[0]] + gamma * max(Q[0, action[0]], Q[initial_state, action[1]])
    Q[i, action[1]] +=  R1[i, action[1]] + gamma * max(Q[0, action[0]], Q[initial_state, action[1]])
  else:
    # if(prevt>= datetime.datetime.timestamp(pd.to_datetime(df.loc[start,'Timestamp']))):
    #   t=1;
    R1 = RMatrix(1, t)
   
    i=0;
    
    Q[i, action[0]] +=  R[i, action[0]] + gamma * max(Q[i+1, action[0]], Q[i+1, action[1]])
    Q[i, action[1]] +=  R1[i, action[1]] + gamma * max(Q[i+1, action[0]], Q[i+1, action[1]])

    i = i + 1
    Q[i, action[0]] +=  R1[i, action[0]] + gamma * max(Q[0, action[0]], Q[initial_state, action[1]])
    Q[i, action[1]] +=  R1[i, action[1]] + gamma * max(Q[0, action[0]], Q[initial_state, action[1]])
    
    
  prev=df.loc[start,'Hash'] 
  date_time_str=df.loc[start,'Timestamp']
  date_time_obj =pd.to_datetime(date_time_str)
  prevt=datetime.datetime.timestamp(date_time_obj) 
  start = start+1
  print (Q, "\n")

 
  
    
     
```
